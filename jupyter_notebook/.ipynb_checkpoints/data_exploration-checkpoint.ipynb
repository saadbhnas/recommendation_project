```python
def get_genres(x):
    
    if 'Action' in x:
        
        return 'Action'
    
    elif 'Comedy' in x:
        
        return 'Comedy'
    
    elif 'Adventure' in x:
        
        return 'Adventure'
    
    elif 'Crime' in x:
        
        return 'Crime'
    
    elif 'Thriller' in x:
        
        return 'Thriller'
    
    elif 'Romance' in x:
        
        return 'Romance'
    
    elif 'War' in x:
        
        return 'war'
    
    elif 'History' in x:
        
        return 'History'
    
    else:
        return 'others'
```

```python
genres_study['genres_bines'] = genres_study['genres'].apply(lambda x : get_genres(x))
```

```python
def tagline_important_words(x):
    if type(x) == str:
        x = x.lower()
        if 'fight' in x:
            return 'fight'
        elif 'love' in x:
            return 'love'
        elif 'friends' in x:
            return 'frineds'
        elif 'family' in x:
            return 'family'
        elif 'surprise' in x:
            return 'surprise'
        elif 'crime' in x:
            return 'crime'
        elif 'power' in x:
            return 'power'
        elif 'hero' in x:
            return 'hero'
        else:
            return 'others'
        
    
    
```

```python
def handle_json_better(x):
    
    #number_of_nones = re.findall(r'\bNone\b',x)
    
    #for i in range(len(number_of_nones)):
    #    match = re.search(r'\bNone\b',x)
    #    if match :
    #        x = x.replace('None' , '"missing"')
    
    try :
        
        x = x.replace("'",'"')
    
        x = json.loads(x)
        
    except :
        
        x = np.nan
    return x
```

```python
#so we have json format that we need to apply handle json function but without the cnversion of non to missing becuase its already done 
#after that we need to target name of the first order 
def get_main_cast_names(x):
    if x == []:
        return np.nan
    elif x == '':
        return np.nan
    else :
        if x[0]['name'] == '' :
            return np.nan
        elif x[0]['name']:
            return x[0]['name']
    return x
```

```python
yy.cast_main_gender = yy.cast_main_gender.apply(lambda x : get_main_cast_gender(x))
```

```python
def get_main_cast_gender(x):
    if x == []:
        return np.nan
    elif x == '':
        return np.nan
    else :
        if x[0]['gender'] == 1:
            return 'F'
        elif x[0]['gender'] == 2:
            return 'M'
        else :
            return np.nan
        
    return x
```

```python
def create_scatter(feature1, feature2):

    scat = plt.scatter(x=data[feature1],
                       y=data[feature2])

    plt.xlabel(feature1.capitalize())
    plt.ylabel(feature2.capitalize())
    plt.title('{} vs {}'.format(feature1,feature2))

    plt.show()
```

```python
widgets.interact(create_scatter, feature1=data, feature2=data)
```
